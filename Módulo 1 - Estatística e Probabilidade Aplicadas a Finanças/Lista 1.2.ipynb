## **1.2 - Momentos estatísticos (média, variância, skewness, curtose)**

Imagine que você está observando os retornos diários de um ativo. Em vez de olhar cada ponto isoladamente, você quer **resumir** essa experiência em alguns números que capturam o essencial. Os momentos estatísticos fazem exatamente isso: são "lentes" progressivas que revelam camadas diferentes da mesma distribuição.

O **primeiro momento** (média) responde: "para onde esse ativo tende a ir?" É o centro de gravidade dos retornos, o ponto de equilíbrio, a **esperança matemática** da distribuição. Formalmente, é o valor que minimiza a soma dos desvios quadráticos e representa o retorno esperado de longo prazo.

O **segundo momento** (variância) pergunta: "quão longe do centro os retornos costumam cair?" É a dispersão, a "largura" da experiência. Um ativo com variância alta é inquieto, salta muito; variância baixa sugere comportamento mais previsível. A variância quantifica o risco idiossincrático e é fundamental para diversificação.

O **terceiro momento** (assimetria ou *skewness*) investiga: "os retornos se espalham de forma equilibrada, ou há uma cauda mais pesada?" Retornos simétricos (skewness ~0) são democráticos: ganhos e perdas extremas têm probabilidades parecidas. Skewness negativa indica que quedas grandes são mais frequentes que altas equivalentes, o pesadelo do gestor, porque o risco "dói" mais do que o ganho "compensa".

O **quarto momento** (curtose ou *kurtosis*) questiona: "eventos extremos acontecem mais ou menos do que o esperado em uma distribuição normal?" Curtose alta revela "caudas gordas": crashes e rallies aparecem com frequência incômoda. Curtose baixa sugere que extremos são raros, quase ausentes. A curtose importa pois, assumir normalidade quando há assimetria ou caudas gordas leva a subestimar risco. Ou seja, um portfólio que "parece seguro" na variância pode esconder quedas brutais na skewness. Um backtest que ignora curtose pode tratar um evento de 5 sigmas como "impossível", até ele acontecer na sexta-feira seguinte.

### **Valor Esperado (Média ou Esperança)**

O valor esperado de uma variável aleatória, também chamado de **média** ou **esperança**, representa o resultado "típico" que esperaríamos observar se pudéssemos repetir o experimento aleatório infinitas vezes e calcular a média de todos os resultados.

**Interpretação intuitiva:** Imagine que você está analisando o retorno diário de uma ação. Se o retorno esperado é $0,1\%$, isso significa que, em média, ao longo de muitos dias, esperamos que a ação suba aproximadamente $0,1\%$ por dia. Alguns dias ela subirá mais, outros menos, e alguns dias ela cairá, mas a média de longo prazo tende a ser $0,1\%$.

O valor esperado pode ser pensado como o "centro de gravidade" de uma distribuição de probabilidade. Se imaginarmos a distribuição como uma prancha com pesos distribuídos ao longo dela (onde cada peso representa a probabilidade de um valor ocorrer), o valor esperado seria o ponto de equilíbrio dessa prancha.

**Exemplo numérico:** Considere uma variável aleatória $X$ que representa o retorno de um investimento com a seguinte distribuição:

- Probabilidade de $50\%$ de retorno de $+10\%$
- Probabilidade de $30\%$ de retorno de $-5\%$
- Probabilidade de $20\%$ de retorno de $+15\%$

O valor esperado seria:
$$
E[X] = 0,50 \times 0,10 + 0,30 \times (-0,05) + 0,20 \times 0,15 = 0,05 + (-0,015) + 0,03 = 0,065 = 6,5\%
$$

Isso significa que, em média, esperamos um retorno de $6,5\%$ neste investimento.

#### **Definição Formal**

Seja $X$ uma variável aleatória definida no espaço de probabilidade $(\Omega, \mathcal{F}, P)$. O **valor esperado** (ou esperança matemática) de $X$, denotado por $E[X]$ ou $\mu$, é definido como:

**Caso discreto:** Se $X$ assume valores $x_{1}, x_{2}, x_{3}, ...$ com probabilidades $p_{1}, p_{2}, p_{3}, ...$, então:
$$
E[X] = \sum_{i} x_{i} \cdot p_{i}
$$
desde que a soma convirja absolutamente.

**Caso contínuo:** Se $X$ possui função densidade de probabilidade $f_{X}(x)$, então:
$$
E[X] = \int_{-\infty}^{+\infty} x \cdot f_{X}(x) \, dx
$$
desde que a integral convirja absolutamente.

**Observação importante:** O valor esperado nem sempre existe. Para que $E[X]$ seja bem definido, é necessário que:
$$
E[|X|] = \int_{-\infty}^{+\infty} |x| \cdot f_{X}(x) \, dx < \infty
$$

Existem distribuições (como a distribuição de Cauchy) para as quais o valor esperado não está definido.

#### **Propriedades do Valor Esperado**

O valor esperado possui propriedades matemáticas importantes que facilitam cálculos e análises:

1. **Linearidade:** Para quaisquer variáveis aleatórias $X$ e $Y$ e constantes $a$ e $b$:
$$
E[aX + bY] = a \cdot E[X] + b \cdot E[Y]
$$
   
   Esta propriedade é extremamente útil e **não requer que $X$ e $Y$ sejam independentes**. Ela nos permite calcular o retorno esperado de um portfólio como uma combinação linear dos retornos esperados dos ativos individuais.

2. **Constante:** Se $c$ é uma constante, então $E[c] = c$.

3. **Monotonicidade:** Se $X \leq Y$ (isto é, $X(\omega) \leq Y(\omega)$ para todo $\omega \in \Omega$), então $E[X] \leq E[Y]$.

4. **Valor esperado do produto (caso independente):** Se $X$ e $Y$ são variáveis aleatórias **independentes**, então:
$$
E[X \cdot Y] = E[X] \cdot E[Y]
$$
   
   **Atenção:** Esta propriedade **NÃO** vale em geral se $X$ e $Y$ não forem independentes.

#### **Aplicações em Finanças Quantitativas**

**Retorno Esperado de Ativos:** O valor esperado é utilizado para calcular o retorno médio esperado de um ativo financeiro, sendo um componente fundamental na construção de portfólios e na avaliação de investimentos.

**Teoria Moderna de Portfólios:** Na construção de portfólios ótimos (como no modelo de Markowitz), o retorno esperado do portfólio é dado pela média ponderada dos retornos esperados dos ativos individuais, usando a propriedade de linearidade.

**Precificação de Derivativos:** Em modelos de precificação como Black-Scholes, o valor esperado aparece no cálculo do preço justo de opções sob uma medida de probabilidade neutra ao risco.

**Análise de Risco-Retorno:** O valor esperado representa o "prêmio" que um investidor espera receber ao assumir determinado risco, sendo comparado com taxas livres de risco para avaliar a atratividade de investimentos.


### **Variância**

A variância é uma medida de dispersão que quantifica o quão distantes os valores de uma variável aleatória tendem a estar de sua média. **Em termos simples, a variância nos diz o quão "espalhados" estão os possíveis resultados.**

**Interpretação intuitiva:** Imagine duas ações com o mesmo retorno esperado de $5\%$ ao ano. A primeira ação tem retornos que oscilam entre $4\%$ e $6\%$, enquanto a segunda oscila entre $-20\%$ e $+30\%$. Claramente, a segunda ação é muito mais arriscada, pois seus retornos variam muito mais. A variância captura exatamente essa diferença: ela será muito maior para a segunda ação.

Em finanças, a variância (ou seu desvio padrão, que é a raiz quadrada da variância) é frequentemente usada como medida de risco. Um ativo com alta variância é considerado mais arriscado porque seus retornos são menos previsíveis.

**Exemplo numérico:** Considere novamente a variável aleatória $X$ do exemplo anterior, com retornos de $+10\%$, $-5\%$ e $+15\%$ com probabilidades $50\%$, $30\%$ e $20\%$, respectivamente. Já calculamos que $E[X] = 6,5\%$.

A variância seria:
$$
\begin{align}
Var(X) &= 0,50 \times (0,10 - 0,065)^{2} + 0,30 \times (-0,05 - 0,065)^{2} + 0,20 \times (0,15 - 0,065)^{2} \\
&= 0,50 \times (0,035)^{2} + 0,30 \times (-0,115)^{2} + 0,20 \times (0,085)^{2} \\
&= 0,00061 + 0,00397 + 0,00145 \\
&= 0,00603
\end{align}
$$

O desvio padrão seria $\sigma = \sqrt{0,00603} \approx 0,0777 = 7,77\%$.

#### **Definição Formal**

Seja $X$ uma variável aleatória com valor esperado $\mu = E[X]$. A **variância** de $X$, denotada por $Var(X)$ ou $\sigma^{2}$, é definida como:
$$
Var(X) = E[(X - \mu)^{2}]
$$

Essa definição nos diz que a variância é o valor esperado dos desvios quadrados em relação à média. Elevamos ao quadrado para evitar que desvios positivos e negativos se cancelem e também para dar mais peso a desvios maiores.

**Fórmula alternativa (computacionalmente útil):**
$$
Var(X) = E[X^{2}] - (E[X])^{2}
$$

Esta fórmula é frequentemente mais conveniente para cálculos, pois não requer subtrair a média de cada observação antes de elevar ao quadrado.

**Desvio Padrão:** O desvio padrão é simplesmente a raiz quadrada da variância:
$$
\sigma = \sqrt{Var(X)}
$$

O desvio padrão tem a vantagem de estar na mesma unidade da variável original, tornando sua interpretação mais direta.

#### **Propriedades da Variância**

A variância possui propriedades importantes, embora seja menos "bem-comportada" que o valor esperado:

1. **Não-negatividade:** $Var(X) \geq 0$ para qualquer variável aleatória $X$, e $Var(X) = 0$ se e somente se $X$ é constante (quase certamente).

2. **Constante:** Se $c$ é uma constante, então $Var(c) = 0$.

3. **Multiplicação por constante:** Para qualquer constante $a$:
$$
Var(aX) = a^{2} \cdot Var(X)
$$
   
   - Observe que a variância **não** é linear na multiplicação por constantes - ela escala com o quadrado da constante.

4. **Adição de constante:** Para qualquer constante $b$:
$$
Var(X + b) = Var(X)
$$
   
   - Adicionar uma constante apenas desloca a distribuição, mas não altera sua dispersão.

5. **Soma de variáveis independentes:** Se $X$ e $Y$ são variáveis aleatórias **independentes**, então:
$$
Var(X + Y) = Var(X) + Var(Y)
$$
   
   - **Caso geral (com covariância):** Se $X$ e $Y$ não são independentes:
$$
Var(X + Y) = Var(X) + Var(Y) + 2 \cdot Cov(X, Y)
$$
   
  - onde $Cov(X, Y) = E[(X - E[X])(Y - E[Y])]$ é a covariância entre $X$ e $Y$.

6. **Desigualdade de Chebyshev:** Para qualquer $k > 0$:
$$
P(|X - \mu| \geq k) \leq \frac{Var(X)}{k^{2}}
$$
   
   - Esta desigualdade nos fornece um limite superior para a probabilidade de a variável se desviar significativamente de sua média.

##### **Aplicações em Finanças Quantitativas**

**Medida de Risco:** A variância (ou volatilidade, que é o desvio padrão) é a medida de risco mais utilizada em finanças. Um ativo com alta volatilidade é considerado mais arriscado.

**Teoria de Markowitz:** A construção de portfólios eficientes busca minimizar a variância do portfólio para um dado nível de retorno esperado. A variância de um portfólio depende não apenas das variâncias individuais dos ativos, mas também de suas covariâncias.

**Modelo CAPM:** O Capital Asset Pricing Model relaciona o retorno esperado de um ativo com sua covariância em relação ao portfólio de mercado, normalizada pela variância do mercado (o beta).

**Precificação de Opções:** A volatilidade (desvio padrão) é um input fundamental em modelos de precificação de opções como Black-Scholes. Opções são mais valiosas quando o ativo subjacente tem maior volatilidade.

**Gestão de Risco:** Medidas como VaR (Value at Risk) e CVaR (Conditional Value at Risk) dependem da estimação da variância e de outros momentos da distribuição de retornos.

### **Assimetria (Skewness)**

A assimetria, ou skewness, mede o grau de assimetria de uma distribuição de probabilidade em relação à sua média. **Enquanto a média nos diz "onde" está o centro e a variância nos diz "quão espalhados" estão os dados, a skewness nos diz se há mais valores extremos de um lado do que do outro.**

**Interpretação intuitiva:** Imagine uma distribuição de retornos de um ativo. Se a distribuição for perfeitamente simétrica (como uma distribuição normal), a skewness será zero - há a mesma probabilidade de desvios positivos e negativos da média.

- **Skewness positiva (assimetria à direita):** A distribuição tem uma "cauda" mais longa à direita. Isso significa que, embora a maioria dos valores esteja concentrada à esquerda, há alguns valores muito grandes à direita. Em finanças, uma skewness positiva sugere que há maior probabilidade de retornos extremamente positivos (bom para investidores).

- **Skewness negativa (assimetria à esquerda):** A distribuição tem uma "cauda" mais longa à esquerda. Isso indica maior probabilidade de perdas extremas. Em finanças, skewness negativa é geralmente indesejável, pois sugere risco de quedas bruscas.

**Exemplo financeiro:** Retornos de ações individuais frequentemente apresentam skewness negativa - há mais probabilidade de grandes quedas (crashes) do que de grandes altas. Por outro lado, estratégias de compra de opções de compra (call options) podem ter skewness positiva.

#### **Definição Formal**

Seja $X$ uma variável aleatória com média $\mu = E[X]$ e desvio padrão $\sigma = \sqrt{Var(X)}$. A **assimetria** (ou coeficiente de assimetria) de $X$, denotada por $Skew(X)$ ou $\gamma_{1}$, é definida como:
$$
Skew(X) = E\left[\left(\frac{X - \mu}{\sigma}\right)^{3}\right] = \frac{E[(X - \mu)^{3}]}{(\sigma)^{3}}
$$

Essa definição padroniza o terceiro momento central dividindo-o pelo cubo do desvio padrão, tornando a medida adimensional e comparável entre diferentes distribuições.

**Fórmula alternativa:**
$$
Skew(X) = \frac{E[X^{3}] - 3\mu\sigma^{2} - \mu^{3}}{\sigma^{3}}
$$

**Interpretação dos valores:**
- $Skew(X) = 0$: distribuição simétrica
- $Skew(X) > 0$: assimetria positiva (cauda à direita)
- $Skew(X) < 0$: assimetria negativa (cauda à esquerda)

Para a distribuição normal, $Skew(X) = 0$. Desvios significativos de zero indicam assimetria.


#### **Propriedades da Assimetria**

1. **Invariância por translação:** Para qualquer constante $b$:
$$
Skew(X + b) = Skew(X)
$$
   
   Adicionar uma constante não altera a forma da distribuição.

2. **Inversão de sinal:** Para qualquer constante $a \neq 0$:
$$
Skew(aX) = \begin{cases}
Skew(X) & \text{se } a > 0 \\
-Skew(X) & \text{se } a < 0
\end{cases}
$$
   
   Multiplicar por uma constante negativa inverte a direção da assimetria.

3. **Distribuições simétricas:** Se $X$ tem distribuição simétrica em torno de sua média, então $Skew(X) = 0$ (mas a recíproca não é necessariamente verdadeira).

4. **Não-aditividade:** Ao contrário da média e variância, a skewness de uma soma de variáveis aleatórias **não** possui uma fórmula simples em termos das skewnesses individuais.



#### **Aplicações em Finanças Quantitativas**



**Análise de Risco:** Investidores aversos ao risco preferem distribuições com skewness positiva, pois oferecem proteção contra perdas extremas e possibilidade de ganhos excepcionais.

**Precificação de Ativos:** Modelos que consideram apenas média e variância (como CAPM tradicional) podem ser inadequados quando há assimetria significativa. Extensões como o modelo de três momentos incorporam a skewness.

**Estratégias de Derivativos:** A skewness implícita nas opções (volatility skew ou smile) é crucial para precificação. Opções de venda (puts) frequentemente têm volatilidade implícita maior que opções de compra (calls) para o mesmo vencimento, refletindo a skewness negativa dos retornos de ações.

**Gestão de Portfólios:** A construção de portfólios pode buscar não apenas maximizar retorno e minimizar variância, mas também maximizar skewness, criando distribuições de retorno mais favoráveis.

**Detecção de Bolhas e Crashes:** Mudanças na skewness dos retornos podem sinalizar regimes de mercado diferentes, ajudando na identificação de períodos de alta volatilidade ou comportamento anômalo.

### **Curtose (Kurtosis)**

A curtose mede a "concentração" de probabilidade nas caudas de uma distribuição em relação a uma distribuição normal. **Em termos simples, a curtose nos diz com que frequência observamos valores extremos (muito distantes da média).**

**Interpretação intuitiva:** Imagine novamente duas distribuições de retornos com a mesma média e variância. A primeira segue uma distribuição normal, onde eventos extremos são raros. A segunda tem mais probabilidade de gerar retornos muito distantes da média - tanto grandes ganhos quanto grandes perdas. A curtose captura essa diferença.

Uma alta curtose indica "caudas pesadas" (fat tails), ou seja, eventos extremos são mais prováveis do que o previsto por uma distribuição normal. Isso é extremamente importante em finanças, pois eventos raros mas extremos (como o crash de 1987 ou a crise de 2008) têm impacto desproporcional.

**Exemplo financeiro:** Retornos de mercados financeiros tipicamente apresentam curtose maior que a distribuição normal. Isso significa que crashes e rallies extremos ocorrem com mais frequência do que modelos baseados na normalidade preveriam.

#### **Definição Formal**

Seja $X$ uma variável aleatória com média $\mu = E[X]$ e desvio padrão $\sigma = \sqrt{Var(X)}$. A **curtose** de $X$, denotada por $Kurt(X)$ ou $\gamma_{2}$, é definida como:
$$
Kurt(X) = E\left[\left(\frac{X - \mu}{\sigma}\right)^{4}\right] = \frac{E[(X - \mu)^{4}]}{(\sigma)^{4}}
$$

Essa é a curtose "não ajustada". Frequentemente, usa-se a **curtose em excesso** (excess kurtosis), definida como:
$$
Kurt_{excess}(X) = Kurt(X) - 3
$$

A subtração de $3$ é feita porque a distribuição normal tem curtose igual a $3$. Dessa forma:
- $Kurt_{excess}(X) = 0$: caudas similares à distribuição normal (mesocúrtica)
- $Kurt_{excess}(X) > 0$: caudas mais pesadas que a normal (leptocúrtica)
- $Kurt_{excess}(X) < 0$: caudas mais leves que a normal (platicúrtica)

**Fórmula alternativa:**
$$
Kurt(X) = \frac{E[X^{4}] - 4\mu E[X^{3}] + 6\mu^{2}E[X^{2}] - 3\mu^{4}}{\sigma^{4}}
$$

#### **Propriedades da Curtose**

1. **Invariância por translação:** Para qualquer constante $b$:
$$
Kurt(X + b) = Kurt(X)
$$

2. **Invariância por mudança de escala:** Para qualquer constante $a \neq 0$:
$$
Kurt(aX) = Kurt(X)
$$
   
   - A curtose é completamente invariante a transformações lineares (desde que $a \neq 0$).

3. **Limite inferior:** Para qualquer variável aleatória, $Kurt(X) \geq 1$, com igualdade se e somente se $X$ assume apenas dois valores distintos.

4. **Distribuição normal:** Para uma distribuição normal, $Kurt(X) = 3$ (ou $Kurt_{excess}(X) = 0$).

5. **Não-aditividade:** Assim como a skewness, a curtose de uma soma de variáveis aleatórias não possui uma fórmula simples em termos das curtoses individuais.



#### **Aplicações em Finanças Quantitativas**



**Gestão de Risco Extremo:** A curtose é essencial para modelar eventos de cauda (tail events). Modelos que ignoram a curtose elevada tendem a subestimar o risco de crashes e crises.

**Value at Risk (VaR) e CVaR:** Estimativas de VaR e CVaR baseadas na normalidade podem ser severamente inadequadas quando a curtose é alta. Métodos que incorporam curtose (como modelos GARCH, distribuições t de Student, ou abordagens de valores extremos) são mais precisos.

**Precificação de Opções:** A curtose implícita afeta a precificação de opções, especialmente opções fora do dinheiro (out-of-the-money). O fenômeno do "volatility smile" está relacionado à curtose e skewness dos retornos.

**Teste de Normalidade:** A curtose, junto com a skewness, é usada em testes estatísticos (como o teste de Jarque-Bera) para verificar se uma série de retornos segue uma distribuição normal.

**Alocação de Ativos:** Investidores aversos ao risco de cauda preferem ativos com menor curtose, pois oferecem menor probabilidade de perdas extremas.

**Estratégias de Hedge:** Entender a curtose ajuda a desenhar estratégias de proteção contra eventos extremos, como o uso de opções de venda para proteção de portfólio.

### **Relação entre os Momentos**

Os quatro primeiros momentos fornecem uma caracterização progressivamente mais completa de uma distribuição de probabilidade:

1. **Média:** Localização central
2. **Variância:** Dispersão em torno da média
3. **Skewness:** Assimetria da distribuição
4. **Curtose:** Probabilidade de eventos extremos

**Em finanças quantitativas, análises sofisticadas frequentemente consideram todos esses momentos simultaneamente.** Por exemplo:

- Um investidor pode buscar maximizar o retorno esperado (média)
- Minimizar o risco (variância)
- Preferir assimetria positiva (skewness > 0)
- Evitar caudas pesadas (curtose baixa)

Essas preferências podem estar em conflito, e encontrar o equilíbrio ótimo é um problema fundamental em gestão de portfólios.

### **Momentos Superiores e Limitações**

Embora os quatro primeiros momentos sejam os mais utilizados, existem momentos de ordem superior que podem fornecer informações adicionais sobre a distribuição. No entanto, na prática:

- **Momentos superiores são progressivamente mais difíceis de estimar com precisão** a partir de dados finitos
- **Eles se tornam cada vez mais sensíveis a outliers**
- **Sua interpretação intuitiva é mais difícil**

Por essas razões, análises práticas em finanças raramente vão além do quarto momento (curtose).

**Limitação importante:** Mesmo conhecendo todos os momentos de uma distribuição, nem sempre é possível reconstruir completamente a distribuição original. No entanto, para muitas distribuições encontradas em finanças, os quatro primeiros momentos fornecem uma caracterização suficientemente boa para fins práticos.

### **Considerações Finais**



Os momentos estatísticos são ferramentas poderosas, mas é importante lembrar suas limitações:

1. **Momentos amostrais vs. populacionais:** Na prática, trabalhamos com estimativas amostrais dos momentos, que têm incerteza associada.

2. **Estabilidade temporal:** Em finanças, momentos não são constantes no tempo. Volatilidade, por exemplo, muda ao longo do tempo (heterocedasticidade).

3. **Dependência:** Momentos sozinhos não capturam toda a estrutura de dependência entre variáveis ou ao longo do tempo.

4. **Não-normalidade:** Muitos modelos financeiros clássicos assumem normalidade, mas como vamos ver, retornos financeiros frequentemente violam essa suposição via skewness e curtose elevadas.

**Dominar o conceito de momentos estatísticos e saber como aplicá-los corretamente é fundamental para qualquer profissional de finanças quantitativas.** Eles formam a base para modelos mais sofisticados e para a compreensão profunda dos riscos e oportunidades presentes nos mercados financeiros.

#### **Exercício 7 - Variância de Portfólio com Correlação**

Um investidor possui dois ativos em carteira:

**Ativo 1:**
- Peso no portfólio: 60%
- Desvio padrão: 20%

**Ativo 2:**
- Peso no portfólio: 40%
- Desvio padrão: 30%

**a)** Se os ativos forem perfeitamente independentes (correlação = 0), calcule a variância e o desvio padrão do portfólio.

**b)** Se os ativos forem perfeitamente correlacionados (correlação = 1), qual seria o desvio padrão do portfólio? (Dica: use que Cov(X,Y) = ρ·σ_X·σ_Y)

**c)** Se os ativos forem negativamente correlacionados (correlação = -0,5), qual seria o desvio padrão do portfólio?

**d)** Compare os três resultados. O que isso nos ensina sobre os benefícios da diversificação?

**e)** Um investidor ingênuo calcula o "risco médio" do portfólio como 0,6 × 20% + 0,4 × 30% = 24%. Por que essa abordagem está incorreta? O risco real (calculado no item a) é maior ou menor? Por quê?


````python
import math

print("Variância de portfólio: Var(wX + (1-w)Y) = w²Var(X) + (1-w)²Var(Y) + 2w(1-w)Cov(X,Y)\n")
print("a) Var(P) = (((0.6)**2)*((0.2)**2) + ((0.4)**2)*((0.3)**2) + 0)")
Var_P = (((0.6)**2)*((0.2)**2) + ((0.4)**2)*((0.3)**2) + 0)
std_P = math.sqrt(Var_P)
print(f"A variância {Var_P:.4f} e o desvio padrão {std_P:.2%}.\n")

print("b) Covariância(X,Y) = Correlação(X,Y) × σ_X × σ_Y")
print("Covariância(X,Y) = 1 × 0.2 × 0.3")
cov = 1*0.2*0.3
print("(((0.6)**2)*((0.2)**2) +  ((0.4)**2) * ((0,3)**2) + (2*0.6*0.4*cov))")
var_P2 = (((0.6)**2)*(0.2)**2) +  ((0.4)**2) * ((0.3)**2) + (2*0.6*0.4*cov)
std_P2 = math.sqrt(var_P2)
print(f"O desvio padrão seria: {std_P2:.2%}\n")
print("c) Covariância(X,Y) = -0.5 × 0.2 × 0.3")
cov2 = -0.5*0.2*0.3
print("(((0.6)**2)*((0.2)**2) +  ((0.4)**2) * ((0,3)**2) + (2*0.6*0.4*cov))")
var_P3 = (((0.6)**2)*((0.2)**2) +  ((0.4)**2) * ((0.3)**2) + (2*0.6*0.4*cov2))
std_P3 = math.sqrt(var_P3)
print(f"O desvio padrão seria: {std_P3:.2%}\n")
print(f"Primeiro resultado: {std_P:.2%} --> risco reduzido;")
print(f"Segundo resultado: {std_P2:.2%} --> risco máximo, sem diversificação;")
print(f"Terceiro resultado: {std_P3:.2%} --> risco drasticamente reduzido;")
print("Isso mostra que a diversificação (combinar ativos com correlação menor que 1) reduz o risco total do portfólio.")
print("Quanto menor a correlação, maior o benefício da diversificação, pois os ativos não se movem juntos (e no caso negativo, eles se movem em oposição, se 'protegendo').\n")
print("e) Esse cálculo só funciona se a correlação for 1, porque assumiria que os ativos se movem juntos, anulando a diversificação.")
print("O risco real é menor. A abordagem desse enunciado não está correta, porque o desvio padrão não é linear. O cálculo ignora a covriância.")
````

#### **Dicas**



**Dica geral:** Sempre que possível, use as propriedades dos momentos para simplificar cálculos. Lembre-se:
- Valor esperado é linear: E[aX + bY] = aE[X] + bE[Y]
- Variância não é linear: Var(aX) = a²Var(X)
- Skewness e curtose mudam ou se mantêm dependendo da transformação

**Para verificação:** 
- Índice de Sharpe = (Retorno - Taxa livre de risco) / Volatilidade
- Covariância(X,Y) = Correlação(X,Y) × σ_X × σ_Y
- Variância de portfólio: Var(wX + (1-w)Y) = w²Var(X) + (1-w)²Var(Y) + 2w(1-w)Cov(X,Y)

**Interpretações importantes:**
- Skewness negativa → mais perdas extremas que ganhos extremos
- Curtose > 3 → mais eventos extremos que distribuição normal
- Correlação < 1 → benefícios da diversificação
